````python
import os
from datetime import datetime

########################################################################################################
#  Last Edited : Rodrigo Rios - 05/04/2018
#
# Function: get_power_consumption_server (time, format=csv, Application/Benchmark string )
# Parameters: (Time in Seconds, output format = csv(default),tab,pipe none will assume CSV Application/Benchmark String)
# Application/Benchmark: Stress-ng command line -  planned: Add other benchmark command lines 
#
# Description: Collect Server power comsumption from Dells racadm tool. By  dispatching an energy collector Job and an Application/Benchmarks
# Function level: Receives (cpu time,cpu cores,log-level[1,0]) as input and dispatch both Benchmark and Collector and generate output CSV Logs
# Sample Output Logs Files: stress_logs-2018-04-02T10:59:12.log  - 
#                           energy_Collector-2018-04-02T10:59:12.log
#
#######################################################################################################

def get_power_consumption_server( Collect_Time_Interval , Output_Format , Application_String):
    String_Assembly_Run_Stress = Application_String  + " >>stress_logs-" + datetime.now().isoformat(timespec='seconds') + ".log &" 
    String_Assembly_Run_Energy_Collector = "sudo nohup ./Dispatch_Racadm_Collector.sh " + str(Collect_Time_Interval) + " " + str(Output_Format) + " >>energy_Collector-" + datetime.now().isoformat(timespec='seconds') + ".log &" 
    print("String montada =", String_Assembly_Run_Stress)
    print("String montada2 =", String_Assembly_Run_Energy_Collector)
    os.system(String_Assembly_Run_Energy_Collector)
    os.system(String_Assembly_Run_Stress)
    
   
#Collector Time, may or not match 
#Run Small CPU Processing
#get_power_consumption_server( 5, "csv" , "sudo stress --cpu 8 --timeout 5s")

#Run Medium CPU Processing, 5min, 16 CPU
#get_power_consumption_server( 300, "csv" , "sudo stress --cpu 16 -v --timeout 300s")

#Run Large CPU Processing
#get_power_consumption_server( 100, "csv" , "sudo stress --cpu 32 -v --timeout 300s")
````

````python
import os
from datetime import datetime
import pandas as pd 

########################################################################################################
#  Last Edited : Rodrigo Rios - 05/04/2018
#
# Function: get_power_consumption_CPU_RAM (time, Application/Benchmark string, FileName )
# Parameters: (Time in Seconds, String, File Output)
# Application/Benchmark: Stress-ng command line -  planned: Add other benchmark command lines 
#
#######################################################################################################

def get_power_consumption_CPU_RAM( Collect_Time_Interval,Application_String, FileName):
    String_Assembly_Run_Stress = Application_String  + " >> stress_logs-" + datetime.now().isoformat(timespec='seconds') + ".log &" 
    String_Assembly_Run_CPU_RAM_Energy_Collector= "sudo nohup pcm.x 1 -i=" + str(Collect_Time_Interval) + " -csv=" + str(FileName) + ".csv 2> /dev/null &"
    #print("String montada = ", String_Assembly_Run_Stress)
    #print("String montada3= ", String_Assembly_Run_CPU_RAM_Energy_Collector)
    os.system(String_Assembly_Run_CPU_RAM_Energy_Collector)
    os.system(String_Assembly_Run_Stress)
      
# Return energy CPU sockets (SKT0, SKT1) joules. And DRAM values: SKT0.1, SKT1.1
def load_filtered_energy_CPU(inputfile):
    csv = pd.read_csv(inputfile, sep=";", skiprows=1)
    # TODO: implement multi-index version
    #csv = pd.read_csv("energy_cpu_ram.csv", sep=";", header=[0, 1])
    return csv.iloc[:, 76:80]

# Test
# Generate Log
out_file = "energy_cpu_ram_16_cores"
get_power_consumption_CPU_RAM(120, "sudo nohup stress --cpu 32 --timeout 120s", out_file )
````

````python
#Support Tools for creating Unique Files Names
import tempfile
import itertools as IT
import os

def uniquify(path, sep = ''):
    def name_sequence():
        count = IT.count()
        yield ''
        while True:
            yield '{s}{n:d}'.format(s = sep, n = next(count))
    orig = tempfile._name_sequence 
    with tempfile._once_lock:
        tempfile._name_sequence = name_sequence()
        path = os.path.normpath(path)
        dirname, basename = os.path.split(path)
        filename, ext = os.path.splitext(basename)
        fd, filename = tempfile.mkstemp(dir = dirname, prefix = filename, suffix = ext)
        tempfile._name_sequence = orig
    return filename

print(uniquify('/tmp/file.pdf'))
````
